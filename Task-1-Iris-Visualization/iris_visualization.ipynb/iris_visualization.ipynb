# Iris Dataset Visualization

## Objective
The objective of this project is to explore and visualize the Iris dataset using Python libraries such as Pandas, Matplotlib, and Seaborn.

This project includes:
- Dataset loading and inspection
- Statistical analysis
- Scatter plots
- Histograms
- Box plots
- Data understanding and visualization

```python
# First Notebook Cell

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset('iris')

# Show first 5 rows
print(iris.head())
```

```python
# Second Cell

# Dataset shape
print("Shape of dataset:", iris.shape)

# Column names
print("\nColumn Names:")
print(iris.columns)

# Dataset info
print("\nDataset Info:")
print(iris.info())

# Statistical summary
print("\nStatistical Summary:")
print(iris.describe())
```

```python
# Scatter Plot

sns.scatterplot(
    x='sepal_length',
    y='sepal_width',
    hue='species',
    data=iris
)

plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
# Histogram

iris.hist(figsize=(10, 8))

plt.suptitle("Histogram of Iris Features")

plt.show()
```

```python
# Box Plot

sns.boxplot(data=iris)

plt.title("Box Plot of Iris Features")

plt.show()
```

# Final Insights

- Iris-setosa is clearly separated from the other flower species.
- Petal length and petal width are highly useful for classification.
- Most numerical features show near-normal distribution.
- Some outliers can be observed in sepal width and petal width.
- Data visualization helps understand feature relationships effectively.

# General Health Query Chatbot (Gemini API)

## Objective
To build an AI-powered chatbot that answers general health-related questions using Google Gemini API with prompt engineering and safety constraints.

## Description
This project demonstrates how to integrate a Large Language Model (LLM) into a Python application to create a conversational chatbot. The chatbot is designed to provide safe, general health information without giving medical diagnosis or treatment advice.

## Technologies Used
- Python
- Google Gemini API
- Requests Library

## Model Used
- gemini-2.5-flash

## Key Features
- AI-powered conversational chatbot
- Prompt engineering for safe responses
- Interactive command-line interface
- Real-time API integration

```python
import requests
import json

API_KEY = "AIzaSyAKPaANsrNAkLxOIMRaxEEv0KFpb4L1hS4"
```

```python
def chat_with_ai(prompt):
    url = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-2.5-flash:generateContent?key={API_KEY}"

    headers = {
        "Content-Type": "application/json"
    }

    data = {
        "contents": [
            {
                "parts": [
                    {
                        "text": f"""
You are a helpful and safe medical assistant.
Only give general health information.
Do NOT give diagnosis or emergency treatment.

User question: {prompt}
"""
                    }
                ]
            }
        ]
    }

    response = requests.post(url, headers=headers, json=data)
    return response.json()
```

```python
def get_reply(prompt):
    res = chat_with_ai(prompt)

    try:
        return res["candidates"][0]["content"]["parts"][0]["text"]
    except:
        return res
```

```python
print(get_reply("What causes fever?"))
```

```python
while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit", "stop"]:
        print("Bot: Goodbye!")
        break

    response = get_reply(user_input)
    print("Bot:", response)
```

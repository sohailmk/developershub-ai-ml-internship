```python
from transformers import pipeline
```

```python
chatbot = pipeline("text-generation", model="distilgpt2")
```

```python
def mental_health_bot(user_input):
    prompt = f"""
You are a calm, empathetic mental health support assistant.
Always respond gently, positively, and supportively.

User: {user_input}
Assistant:
"""

    result = chatbot(
        prompt,
        max_new_tokens=120,
        do_sample=True,
        temperature=0.7,
        pad_token_id=50256
    )

    return result[0]["generated_text"]
```

```python
print(mental_health_bot("I feel stressed and overwhelmed with studies"))
```
